```python
t = Test()
numeric_cols = [col for col, dtype in zip(df.columns, df.dtypes) if dtype != 'object']
normal_results = t.applyNormalTest(df[numeric_cols])[1]
normal_features= set()
for key in normal_results:
    normal_features= normal_features | set(normal_results[key])
print(normal_features)
t.applyParametricTest(df[df['Sexo']=="Chico"],df[df['Sexo']=="Chica"],"Hombre","Mujer",list(normal_features))
t.applyNonParametricTest(df[df['Sexo']=="Chico"],df[df['Sexo']=="Chica"],"Hombre","Mujer",numeric_cols)
```

Use of the filter method of a sequence:

  Criteria function

```python
s = SequenceDirectory(os.getcwd()+"/ExampleDirectory")
```

```python
def sequenciasMenores4letras(arraySequencias):
  result=[]
  for i in arraySequencias:
    if len(i.metadata["text"]) < 4:
      result.append(i)
  return result
```

Using de filter

```python
for i in s.filter("files/files",sequenciasMenores4letras):
  print(i)
  print(type(i))
```

Creating a new criteria function for use in filtermetada function:

```python
def sequenciasMenores4letrasMetadata(arraySequencias):
  result=[]
  for i in arraySequencias:
    if len(i) < 4:
      result.append(i)
  return result
```

```python
for i in s.filterMetadata("files/files/text", sequenciasMenores4letrasMetadata):
  print(i)
  print(type(i))
```
